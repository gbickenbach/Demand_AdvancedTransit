```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("drive/MyDrive/FYREE/data.csv", parse_dates=["Date", "Time"])
```

```python
data = data.sort_values("Date")
data["Month"] = data["Date"].dt.month
data["Year"] = data["Date"].dt.year
data = data.reset_index()
data = data.drop(columns = "index")
```

```python
data23f = data
```

```python
#START FROM HERE
```

```python
data23f = data23f.drop(columns="Unnamed: 0")
```

```python

data23f["Day"] =""
```

```python

for i, row in enumerate(data23f.iloc[:, 0]):
  data23f.iloc[i, 11] = data23f["Date"][i].day_name()

```

```python
data23f
```

```python

data23f = data23f.sort_values(by = ['Stop', "Day", "Month"], ascending = [True, True, True], na_position = 'first')
data23f["Passengers"] = data23f["Passengers"].fillna(0)
data23f["Count"] =""
```

```python
data23f
```

```python
for i, row in enumerate(data23f.iloc[1:, 8]):
  if ((data23f.iloc[i, 5] == data23f.iloc[i-1, 5])&(data23f.iloc[i, 11]==data23f.iloc[i-1, 11])):
    data23f.iloc[i,12] = data23f.iloc[i, 8] + data23f.iloc[i-1, 12]
  else:
    data23f.iloc[i,12] = 0
```

```python
data23f
```

```python
february_demand = pd.DataFrame()
for i, row in enumerate(data23f.iloc[:-1, 11]):
  if (row != data23f.iloc[i+1, 11]):
    february_demand = february_demand.append(data23f.iloc[i])
february_demand = february_demand.append(data23f.iloc[-2])
```

```python
february_demand
```

```python
demand_february = pd.DataFrame()
demand_february["Station"] = february_demand.iloc[:,5]
demand_february["Day of the Week"] = february_demand.iloc[:,11]
demand_february["Demand"] = february_demand.iloc[:,12]
demand_february
```

```python
demand_february.to_csv("drive/MyDrive/FYREE/Weekdays/Weekday_Demand.csv")
```
