```python
import pandas as pd
import math
```

```python
data = pd.read_csv("drive/MyDrive/FYREE/Real/Middlebury_Demand.csv")
data = data.drop(columns ="Unnamed: 0")
data["Block Group Work"] = ''
data["Block Group Home"] = ''
```

```python
data
```

```python
for i , row in enumerate(data.iloc[:, 0]):
  data.iloc[i, 23] = math.floor((data.iloc[i, 0]  - 500019608000000) / 1000)
  data.iloc[i, 24] = math.floor((data.iloc[i, 1]  - 500019608000000) / 1000)
```

```python
work = pd.DataFrame()
home = pd.DataFrame()
home = data
work = data
work["Count"] = 0
home["Count"] = 0
```

```python
homegroups = pd.DataFrame()
workgroups = pd.DataFrame()
```

```python
work = work.sort_values(by = ['w_geocode', 'h_geocode'], ascending = [True, True], na_position = 'first')
for i, row in enumerate(work.iloc[1:, 23]):
  if ((work.iloc[i, 23] == work.iloc[i-1, 23])):
    work.iloc[i,25] = work.iloc[i, 2] + work.iloc[i-1, 25]
  else:
    work.iloc[i,25] = 0

for i, row in enumerate(work.iloc[:, 25]):
  if (work.iloc[i, 25] != 0):
    work.iloc[i,25] += 1

work.iloc[-1, 25] = work.iloc[-2, 25] + work.iloc[-1, 2]



for i, row in enumerate(work.iloc[:-1, 23]):
  if (row != work.iloc[i+1, 23]):
    workgroups = workgroups.append(work.iloc[i])
workgroups = workgroups.append(work.iloc[-1])
```

```python
home = home.sort_values(by = ['h_geocode', 'w_geocode'], ascending = [True, True], na_position = 'first')
for i, row in enumerate(home.iloc[1:, 24]):
  if ((home.iloc[i, 24] == home.iloc[i-1, 24])):
    home.iloc[i,25] = home.iloc[i, 2] + home.iloc[i-1, 25]
  else:
    home.iloc[i,25] = 0

for i, row in enumerate(home.iloc[:, 25]):
  if (home.iloc[i, 25] != 0):
    home.iloc[i,25] += 1

home.iloc[-1, 25] = home.iloc[-2, 25] + home.iloc[-1, 2]


for i, row in enumerate(home.iloc[:-1, 24]):
  if (row != home.iloc[i+1, 24]):
    homegroups = homegroups.append(home.iloc[i])
homegroups = homegroups.append(home.iloc[-1])
```

```python
wfinal = pd.DataFrame()
hfinal = pd.DataFrame()
```

```python
wfinal["Block Group Work"] = workgroups.iloc[:,23]
wfinal["Work Count"] = workgroups.iloc[:, 25]
hfinal["Block Group Home"] = homegroups.iloc[:, 24]
hfinal["Home Count"] = homegroups.iloc[:, 25]
```

```python
wfinal.to_csv('drive/MyDrive/FYREE/Part Two/Block Grouping/work.csv')
hfinal.to_csv('drive/MyDrive/FYREE/Part Two/Block Grouping/home.csv')
```
